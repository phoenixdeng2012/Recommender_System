# Recommender system using AutoEncoder

```python
from MovieLens import MovieLens
from AutoRecAlgorithm import AutoRecAlgorithm
from Evaluator import Evaluator
```

```python
import random
import numpy as np
```

```python
def LoadMovieLensData():
    ml=MovieLens()
    data=ml.loadMovieLensLatestSmall()
    rankings=ml.getPopularityRanks()
    return (ml, data, rankings)
```

```python
np.random.seed(0)
random.seed(0)
```

```python
# load data set for the recommender algorithms
ml,data, rankings=LoadMovieLensData()
```

```python
# construct an Evaluator to evaluate them
evaluator=Evaluator(data, rankings)
```

```python
#Autoencoder
AutoRec=AutoRecAlgorithm()
evaluator.AddAlgorithm(AutoRec, 'AutoRec')
```

```python
evaluator.Evaluate(True)
```

```python
evaluator.SampleTopNRecs(ml)
```
